```python
statusEnum = ['Foreclosure', 'Active', 'Price Raised', 'Extended', 'Coming soon', 'Townhouse for sale', 'Active/Pending-Call', 'Re-activated', 'Price Change', 'Pending', 'New construction', 'Lot / Land for sale', 'FOR_SALE', 'Temporarily No Showings', 'New', 'Home for sale', 'For sale by owner', 'Price Reduced', 'Back On Market', 'House for sale', 'Multi-family home for sale', 'Auction', 'Condo for sale', 'Bumpable Buyer', 'ACTIVE']
data = []
y = []
with open('./dataset/raw/us-cities-real-estate-sample-zenrows.csv', 'r') as f:
    f.readline()
    for i in range(2,1001):
        d = f.readline().split(',')
        try:
            y.append(float(d[11]))
        except:
            continue
        data.append([statusEnum.index(d[7]), (0 if d[21] == 'NULL' else float(d[21])),(0 if d[22] == 'NULL' else float(d[22])),(0 if d[23] == 'NULL' else float(d[23]))])
print(data)
print(y)
```

```python
training_data = np.array(data[:800])
test_data = np.array(data[799:])
```

# Training Process 

1. **Create a vector of weights for each data point.**
2. **Calculate the dot product of the weights and the data points.**
3. **Calculate the error.**
4. **Calculate the gradient.**
5. **Update the weights.**
6. **Repeat the process.**


```python
weights  =  np.array([0.1, 0.1, 0.1, 0.1])   
print(weights)   
```

```python
# stochastic gradient descent#check if d[11] can be used as target variable by checking if its converted to float
weights  =  np.array([0.1, 0.1, 0.1, 0.1])   
learning_rate = 3
for i in range(0,800):
    predicate =  np.dot(weights, training_data[i])
    loss = y[i] - predicate     
    print(loss)
    g_loss_function =learning_rate * loss * training_data[i]
    weights = weights + g_loss_function
    
```
